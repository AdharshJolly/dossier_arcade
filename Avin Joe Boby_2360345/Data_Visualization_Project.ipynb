```python
import pandas as pd
file_path = '/content/climate_change_impact_on_agriculture_2024.csv'
data = pd.read_csv(file_path)
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

# Average Temperature over the Years
plt.figure(figsize=(10, 6))
sns.lineplot(x="Year", y="Average_Temperature_C", data=data, hue="Country", marker="o")
plt.title('Average Temperature Over the Years by Country')
plt.xlabel('Year')
plt.ylabel('Average Temperature (°C)')
plt.xticks(rotation=45)
plt.legend(loc='best')
plt.tight_layout()
plt.show()

```

```python
#Total Precipitation by Region
plt.figure(figsize=(10, 6))
sns.boxplot(x="Region", y="Total_Precipitation_mm", data=data)
plt.title('Total Precipitation by Region')
plt.xlabel('Region')
plt.ylabel('Total Precipitation (mm)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Crop Yield vs CO2 Emissions
plt.figure(figsize=(10, 6))
sns.scatterplot(x="CO2_Emissions_MT", y="Crop_Yield_MT_per_HA", hue="Crop_Type", data=data, s=100)
plt.title('Crop Yield vs CO2 Emissions by Crop Type')
plt.xlabel('CO2 Emissions (MT)')
plt.ylabel('Crop Yield (MT/HA)')
plt.tight_layout()
plt.show()
```

```python
# Economic Impact of Extreme Weather Events
plt.figure(figsize=(10, 6))
sns.barplot(x="Extreme_Weather_Events", y="Economic_Impact_Million_USD", hue="Country", data=data)
plt.title('Economic Impact of Extreme Weather Events by Country')
plt.xlabel('Extreme Weather Events')
plt.ylabel('Economic Impact (Million USD)')
plt.tight_layout()
plt.show()
```

```python
# Irrigation Access by Crop Type
plt.figure(figsize=(10, 6))
sns.barplot(x="Crop_Type", y="Irrigation_Access_%", data=data, palette="muted")
plt.title('Irrigation Access by Crop Type')
plt.xlabel('Crop Type')
plt.ylabel('Irrigation Access (%)')
plt.tight_layout()
plt.show()
```

```python
# Crop Type Distribution
strategy_counts = data["Crop_Type"].value_counts()
plt.figure(figsize=(10, 8))
plt.pie(strategy_counts,labels=strategy_counts.index,autopct="%1.1f%%",colors=sns.color_palette("Set2"),)
plt.title("Distribution of Crop Types")
plt.tight_layout()
plt.show()
```
